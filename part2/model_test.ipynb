### Evaluator

```python
prompt = f"""
You are an expert ML examiner.

Question: {question["question"]}
Reference answer: {question["target_answer"]}
Student answer: {student_answer}

Evaluate the student's answer for correctness, completeness, and precision.
Explain briefly what is missing or incorrect.
Then provide a numeric score from 0 to 100 in the format:
Score: <number>
Feedback: <short explanation>
"""
```

```python
from transformers import pipeline
evaluator = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.2", max_new_tokens=300)

response = evaluator(prompt)[0]["generated_text"]
```
